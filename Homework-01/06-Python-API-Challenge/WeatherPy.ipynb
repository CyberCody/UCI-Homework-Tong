```python
import pandas as pd
import numpy as np
import requests
import json
import random as rd
from citipy import citipy
import matplotlib.pyplot as plt
from scipy.stats import linregress
from sklearn import datasets

# Google API Key
from api_keys import weather_api_key
```

## Generate Cities List

```python
#create list to store cities
cities_list = []



while len(cities_list) < 550:
    
    # randomize lat and long
    latitude = rd.uniform(-90.0,90.0)
    longitude = rd.uniform(-180.0,180.0)
    
    cities = citipy.nearest_city(latitude, longitude).city_name
    
    cities_list.append(cities)
len(cities_list)

cities_list
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# generate url to search cities
url = "http://api.openweathermap.org/data/2.5/weather?"
units = 'imperial'

query_url = f"{url}appid={weather_api_key}&units={units}&q="

found_city = []
lat = []
lng = []
max_temp = []
humidity = []
cloudiness = []
wind_speed = []
country = []
date = []

for city in cities_list:
    
    print(f'Retrieving results for {city}')
    response = requests.get(query_url + city).json()
    
    try:
        print(f'Found {city} appending stats...')
        found_city.append(response['name'])
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon']) 
        max_temp.append(response['main']['temp_max'])
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        wind_speed.append(response['wind']['speed'])
        country.append(response['sys']['country'])
        date.append(response['dt'])
    except(KeyError):
              print(f'{city} results not found. Proceeding to check next city...')
                    
    print('------------------------')
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# create dictionary to hold df
mydict = {
    'City': found_city,
    'Lat': lat,
    'Lng': lng,
    'Max Temp': max_temp,
    'Humidity' : humidity,
    'Cloudiness': cloudiness,
    'Wind Speed': wind_speed,
    'Country': country,
    'Date': date
}

# display df
city_df = pd.DataFrame(mydict)
city_df.to_csv(r'Images/cities.csv')
```

## Linear Regression

```python
# separate southern hemisphere
southern = pd.DataFrame(city_df[city_df['Lat'] < 0])
northern = pd.DataFrame(city_df[city_df['Lat'] > 0])

# check display
southern.head()
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python

# Print out the r-squared value along with the plot.
x_values = northern['Lat']
y_values = northern['Max Temp']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(0,100),fontsize=15,color="red")
plt.xlabel('Latitude')
plt.ylabel('Max Temp')
print(f"The r-squared is: {rvalue**2}")
plt.show()

# analysis
print('Scatter plot shows strong positive corrolation based on the r-value of 0.8144203173760104. Again proving that the close you get to a latitude of 0 the hight the temperature will be.')
```
